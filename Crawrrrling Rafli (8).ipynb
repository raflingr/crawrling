```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import csv
import time
```

```python
#INSTALL BROWSER SUCH CHROME OR EDGE
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
data = []

for index in range (1,4):
    browser.get(f'https://www.tiket.com/homes/villa/indonesia/region/east-java-108001534490276152/page-{index}')
    time.sleep(5)
    soup = BeautifulSoup(browser.page_source,"lxml")
    rows = soup.find_all('div', class_='hotel-card seo-card')
    
    for row in rows:
        nama_villa = row.find(class_='title ellipsis').text
        lokasi_villa = row.find(class_='location ellipsis').text
        rating = row.find(class_='tiket-rating')
        if rating:
            rating = rating.text
        impress = row.find(class_='tiket-impression')
        if impress:
            impress = impress.text
        harga = row.find(class_='after-price')
        if harga:
            harga = harga.text
        deskripsi = row.find(class_='desc-price')
        if deskripsi:
            deskripsi = deskripsi.text
        df = {
            'Name': nama_villa,
            'Location': lokasi_villa,
            'Rating': rating,
            'Impression': impress,
            'Price': harga,
            'Description': deskripsi,
            }
        data.append(df)
```

```python
df = pd.DataFrame(data)
```
